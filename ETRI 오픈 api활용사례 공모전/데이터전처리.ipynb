### DB에서 코스닥 파일가져오기

```python
import sqlite3
import csv
```

```python
# Connect to the SQLite database
connection = sqlite3.connect('data/Daily_Stock_Price.db')

# Create a cursor
cursor = connection.cursor()

# List all tables in the database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name LIKE 'Q%';")
tables = cursor.fetchall()

# Define the folder where you want to save the CSV files
output_folder = 'csv_output/'

# 종목코드
table_nameList = []

# Create CSV files for each table
for table in tables:
    table_name = table[0]
    cursor.execute(f"SELECT * FROM {table_name};")
    data = cursor.fetchall()

    # 종목코드
    table_nameList.append(table_name)
    
    # Define the CSV file path
    csv_file_path = f"{output_folder}{table_name}.csv"

    # Write the data to the CSV file
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        # Write the header (column names)
        csv_writer.writerow([description[0] for description in cursor.description])
        # Write the data rows
        csv_writer.writerows(data)

    print(f"Table '{table_name}' has been saved as '{csv_file_path}'")

# Close the cursor and connection
cursor.close()
connection.close()
```

### 전처리

```python
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm
```

```python
kosdaqList = glob('csv_output/Q' + '*')
```

```python
for _ in tqdm(range(len(kosdaqList))):
    df = pd.read_csv(kosdaqList[_])
    df['code'] = table_nameList[_]

    df['change'] = df.close.diff()/df.close.shift(1)
    df['open'] = df['open'].shift(-1)
    df.dropna(inplace=True)
    df.to_csv(kosdaqList[_], index=False, encoding='utf-8')
```

```python
df = pd.DataFrame([])
for csv_file in tqdm(kosdaqList):
    df = pd.concat([df, pd.read_csv(csv_file)], axis=0)
```

```python
ndf = df.groupby('date').agg(dict(zip(df.columns[1:], [sum]*len(df.columns[1:])))).drop('code',axis=1)
```

```python
ndf.to_csv('코스닥.csv', encoding='utf-8')
```

### DB에서 코스피 파일 가져오기

```python
# Connect to the SQLite database
connection = sqlite3.connect('data/Daily_Stock_Price.db')

# Create a cursor
cursor = connection.cursor()

# List all tables in the database
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name LIKE 'A%';")
tables = cursor.fetchall()

# Define the folder where you want to save the CSV files
output_folder = 'csv_output/'

# 종목코드
table_nameList = []

# Create CSV files for each table
for table in tables:
    table_name = table[0]
    cursor.execute(f"SELECT * FROM {table_name};")
    data = cursor.fetchall()

    # 종목코드
    table_nameList.append(table_name)
    
    # Define the CSV file path
    csv_file_path = f"{output_folder}{table_name}.csv"

    # Write the data to the CSV file
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)
        # Write the header (column names)
        csv_writer.writerow([description[0] for description in cursor.description])
        # Write the data rows
        csv_writer.writerows(data)

    print(f"Table '{table_name}' has been saved as '{csv_file_path}'")

# Close the cursor and connection
cursor.close()
connection.close()
```

```python
kospiList = glob('csv_output/A' + '*')
```

```python
for _ in tqdm(range(len(kospiList))):
    df_ = pd.read_csv(kospiList[_])
    df_['code'] = table_nameList[_]

    df_['change'] = df_.close.diff()/df_.close.shift(1)
    df_['open'] = df_['open'].shift(-1)
    df_.dropna(inplace=True)
    df_.to_csv(kospiList[_], index=False, encoding='utf-8')
```

```python
df_ = pd.DataFrame([])
for csv_file in tqdm(kospiList):
    df_ = pd.concat([df_, pd.read_csv(csv_file)], axis=0)
```

```python
ndf_ = df_.groupby('date').agg(dict(zip(df_.columns[1:], [sum]*len(df_.columns[1:])))).drop('code',axis=1)
ndf_.to_csv('코스피.csv', encoding='utf-8')
```
